```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn import datasets
from sklearn.model_selection import train_test_split
iris = datasets.load_iris(as_frame=True)
X, y =  iris['data'], iris['target']
print(X.head())
print(y.head())
print(X.shape, y.shape)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
#ax.scatter(X.iloc[:, 0], X.iloc[:, 1], X.iloc[:, 2], c=y, cmap='viridis')
sc = ax.scatter(X.iloc[:, 0], X.iloc[:, 1], X.iloc[:, 2], c=y[:],s=50* X.iloc[:, 3], cmap='viridis')
plt.colorbar(sc, ax=ax, label='Target Class')

ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_zlabel(iris.feature_names[2])
ax.set_title('3D Visualization of Iris Dataset')
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_title('PCA of Iris Dataset')
plt.show()
```
