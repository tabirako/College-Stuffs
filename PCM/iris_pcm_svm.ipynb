```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn import datasets
from sklearn.model_selection import train_test_split
iris = datasets.load_iris(as_frame=True)
X, y =  iris['data'], iris['target']
print(X.head())
print(y.head())
print(X.shape, y.shape)
```

```python
fig = plt.figure(figsize=(8, 8))
for feature_a in X.columns:
    for feature_b in X.columns:
        if feature_a == feature_b:
            continue
        ax = fig.add_subplot(4, 4, X.columns.get_loc(feature_a) * 4 + X.columns.get_loc(feature_b) + 1)
        ax.scatter(X[feature_a], X[feature_b], c=y)
        ax.set_xlabel(feature_a)
        ax.set_ylabel(feature_b)

plt.tight_layout()
plt.show()
    
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
#ax.scatter(X.iloc[:, 0], X.iloc[:, 1], X.iloc[:, 2], c=y, cmap='viridis')
sc = ax.scatter(X.iloc[:, 0], X.iloc[:, 1], X.iloc[:, 2], s=50*np.log(X.iloc[:, 3] + 1), c=y[:], cmap='viridis')
    
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_zlabel(iris.feature_names[2])
ax.set_title('3D Visualization of Iris Dataset\n(Point size scaled by ' + iris.feature_names[3] + ')')
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
ax.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_title('PCA of Iris Dataset')
plt.show()
```
